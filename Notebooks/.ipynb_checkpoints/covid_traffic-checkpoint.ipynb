```python
from shapely.geometry import Point
import pandas as pd
import datetime
```

### Imports above, DF creation below

```python
traffic_df = pd.read_csv('../Data/Traffic_Accidents.csv', low_memory=False)
```

```python
covid_df = pd.read_excel('../Data/Covid_county.XLSX')
```

# Traffic cleaning

### Par down to Nashville, Drop unnecessary columns

```python
nash_traffic_df = traffic_df[traffic_df['City'] == 'NASHVILLE']
nash_traffic_df = nash_traffic_df.drop(columns = ['Collision Type Code', 'RPA', 'Precinct', 'Illumination Description', 'Illumination Code', 'Weather Code', 'Weather Description', 'Reporting Officer', 'Mapped Location'])
```

### Creation of month and year columns for sorting

```python
traffic_date = pd.to_datetime(nash_traffic_df['Date and Time'])
nash_traffic_df['Date and Time'] = traffic_date
```

```python
nash_traffic_df['month'] = nash_traffic_df['Date and Time'].dt.month
nash_traffic_df['year'] = nash_traffic_df['Date and Time'].dt.year
```

```python
nash_traffic_df = nash_traffic_df.sort_values(by='Date and Time')
nash_traffic_df
```

### Drop 2013

```python
nash_traffic_df = nash_traffic_df[nash_traffic_df['year'] > 2013]
```

# COVID cleaning

### Par down to Davidson

```python
covid_df = covid_df[covid_df['COUNTY'] == 'Davidson']
```
